```python
import os
import xml.etree.ElementTree as ET

questions = []
for file in os.listdir('../cleaned_eslo_interview/'):
    file_path = os.path.join('../cleaned_eslo_interview/', file)
    # print(file_path)
    root = ET.parse(file_path).getroot()
#     all_descendants = list(root.iter('Turn'))
#     print(all_descendants[30:40])
#     print([elem.tag for elem in root.iter()])
    tags = [t for t in root.iter('Turn')]
#     print(tags)
    for i in range(len(tags)):
        tag = tags[i]
        text = tag.text
#         print(len(text))
        try:
            for t in text.split('\n'):
                if '?' in t and t!='?':
                    questions.append((tags[i-1].text, t))
        except:
            pass
print(len(questions))
# print(questions[0])
```

```python
import pandas as pd
df = pd.DataFrame(questions, columns=['previous_sentence', 'question'])
```

```python
df.to_csv('eslo_interview.csv', sep=';', encoding='cp1252')
```

```python
for file in os.listdir('../eslo_interview/'):
    file_path = os.path.join('../eslo_interview/', file)
    with open(file_path,'r', encoding='cp1252') as f:
        content = f.readlines()
        for i in range(len(content)):
            if content[i].startswith(('<Sync','<Who','<Event')):
                content[i]=''
    with open(f"cleaned_{file_path.split('/')[-1]}",'w', encoding='utf8') as f:
        for c in content:
            f.write(c)
```
